**Data Cleaning**

```python
# Removing duplicates
df=df.drop_duplicates()
df
```

```python
# Null fuction as detecting the missing values
print(df.isnull())
```

```python
# Sum of Columnwise Null  Values
print(df.isnull().sum())
```

```python
 #Drop rows with any missing values
df=df.dropna()
df
```

```python
df=df.info()
df
```

```python
df.describe()
df
```

**Ranking**/**Sorting**

```python
#Totalcase
df=df.sort_values(by="TotalCases")
df
```

```python
#Total cases,newcases,death,Recovered,active,serious_critical
total_cases=df['TotalCases'].sum()
total_newcases=df['NewCases'].sum()
total_death =df['TotalDeaths'].sum()
total_recovered=df['TotalRecovered'].sum()
total_active=df['ActiveCases'].sum()
total_Serious_Critical=df['Serious,Critical'].sum()
total_tests=df['TotalTests'].sum()
total_NewRecovered=df['NewRecovered'].sum()
total_NewDeaths=df['NewDeaths'].sum()
total_Country_Region=df['Country/Region'].sum()
total_population=df['Population'].sum()


print("TotalCases", total_cases)
print("NewCases", total_newcases)
print('Total Death case', total_death)
print('TotalRecovered',total_recovered)
print('ActiveCases',total_active)
print('Serious,Critical',total_Serious_Critical)
print('TotalTests',total_tests)
print('NewRecovered',total_NewRecovered)
print('NewDeaths',total_NewDeaths)
print('Country/Region:',total_Country_Region)
print('Population',total_population)
```

Data Visualization

```python
#Plotting Total case countrywise
plt.style.use('dark_background')
plt.style.use('ggplot')
plt.figure(figsize=(15,4))
plt.plot(df['Country/Region'],df['TotalCases'])
plt.xlabel('Country/Region')
plt.ylabel('TotalCases')
plt.title('TotalCases_countrywise',fontsize=16,fontweight='bold',style='italic')
plt.show()



```

```python
#plotting TotalDeaths
plt.figure(figsize=(8,6))
sns.histplot(x=df['TotalDeaths'],bins=20,kde=True)
plt.title('Distribution of TotalDeaths')
plt.xlabel('TotalDeaths')
plt.ylabel('Frequency')
plt.show()
```

```python
#plotting totalcases vs totaldeaths
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['TotalCases'],y=df['TotalDeaths'])
plt.title('TotalCases_vs_TotalDeaths')
plt.xlabel('TotalCases')
plt.ylabel('TotalDeaths')
plt.show()
```

```python
# Convert columns to numeric type before calculating correlation, errors='coerce' will replace invalid parsing with NaN
plt.figure(figsize=(10, 8))
numeric_df = df.drop(columns=['Country/Region']).apply(pd.to_numeric, errors='coerce')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Exclude the 'Country/Region' column when calculating the median
numeric_df = df.drop(columns=['Country/Region'])

# Select only numeric columns
numeric_df = numeric_df.select_dtypes(include='number')

# Fill NaN values only in numeric columns
df[numeric_df.columns] = df[numeric_df.columns].fillna(numeric_df.median())
df
```

```python
#plt.style.use('ggplot')
plt.figure(figsize=(10,4))
plt.style.use('dark_background') # Changed style name to 'dark_background'

#plotting totalcases
plt.plot(df['Country/Region'],df['TotalCases'],linestyle='-',color='red',label='TotalCases')
#plotting totaldeath
plt.plot(df['Country/Region'],df['TotalDeaths'],linestyle='-',color='blue',label='TotalDeaths')
#plotting total recovered
plt.plot(df['Country/Region'],df['TotalRecovered'],linestyle='-',color='green',label='TotalRecoverd')
#plotting active cases
plt.plot(df['Country/Region'],df['ActiveCases'],linestyle='-',color='yellow',label='ActiveCases')
#plotting serious/critical
#Corrected column name from 'Serious/Critical' to 'Serious,Critical'
plt.plot(df['Country/Region'],df['Serious,Critical'],linestyle='-',color='orange',label='Serious/Critical')
plt.xlabel('Country/Region')
plt.ylabel('No: of cases(in thousand)')
plt.legend()
plt.show()
```

```python
#plotting totalrecovered case vs active cases
plt.figure(figsize=(10,4))
plt.style.use('dark_background')
sns.lineplot(x=df['TotalRecovered'],y=df['ActiveCases'])
plt.show()
```

```python
#Totalcase of WHO reason wise
who=df.groupby('WHO Region')['TotalCases'].sum().sort_values()
print(who)
```

```python
#Data visulization of Totalcases case of WHO Reason
plt.style.use('ggplot')
plt.style.use('dark_background')
plt.figure(figsize=(10,4))
sns.lineplot(x=df['WHO Region'],y=df['TotalCases'],data=df,color='red')
plt.show()
```

```python
# pie chart
plt.figure(figsize=(4,4))

#Defining the variable Who_regionscon
Who_regionscon = df.groupby('WHO Region')['TotalCases'].sum().sort_values() #This assumes that you want to group by TotalCases, you can replace this with the appropriate column

plt.pie(Who_regionscon,labels=Who_regionscon.index,autopct='%1.1f%%',startangle=140,colors=sns.color_palette('pastel'))
plt.title('Totalcase vs WHO Region')
plt.show()
```
